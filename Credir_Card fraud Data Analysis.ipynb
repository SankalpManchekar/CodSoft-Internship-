```python
X=df1.drop(columns='Class',axis=1)
```

```python
Y=df1['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train , Y_train)
```

```python
X_train_prediction=model.predict(X_train)
```

```python
train_data_acc=accuracy_score(X_train_prediction,Y_train)*100
```

```python
print(f"training data accuracy:- {train_data_acc}%")
```

```python
X_test_prediction=model.predict(X_test)
```

```python
test_data_acc=accuracy_score(X_test_prediction,Y_test)*100
```

```python
print(f"Test data accuracy:- {test_data_acc}%")
```
